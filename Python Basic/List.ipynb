## Accessing Items from a list

```python
# Indexing
l = [1,2,3,4,5]
# positive indexing
print(l[0])
#negative indexing
print(l[-3])


# print(l[30]) => list index out of range


# for 2D indexing
l = [1,2,3,[4,5]]  # i want 4
print(l[3][0])
print(l[-1][-2])


# for 3d indexing
l = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(l[0][0][1])
```

```python
# Slicing
l = [1,2,3,4,5,6]

print(l[0:3])
print(l[-3:])
print(l[:-1:2])
print(l[::-1])
```

## Adding Items to a List

* append ==> add single item at last
* extend ==> add multiple item at last
* insert ==> (like a append with particular position) add only one item at a time at certain position in list. use l.insert(index, value) format

```python
# append (add one item at a time at the last of list)
l = [1,2,3,4,5]
# l.append(6,7) TypeError: list.append() takes exactly one argument (2 given)
l.append([6,7])
l.append(10)
print(l)
```

```python
# extend (can add multiple items)
l = [1,2,3,4,5]
l.extend([6,7,8])
l.extend('delhi')
print(l)
```

```python
# insert

l = [1,2,3,4,5]
l.insert(1,'happy')
print(l)
```

## Editing items in a List

```python
l = [1,2,3,4,5]

# editing with indexing
l[-1] = 100
print(l)

# editing with slicing
l[1:-1] = 200,300,400
print(l)
```

## Deleting items from a list (we can delete particular ite)

* del => can delete whole list and particular item using index.
* remove => can delete the item with throwing the item. no need to index.
* pop => delete the last element of list or particular item with giving the index number.
* clear => don't delete the list but remove all items from the list.

```python
# del (have to give index position)
l = [1,2,3,4,5]
# del(l)

# using indexing
del(l[-1])
print(l)

# using slicing
del l[1:3]
print(l)
```

```python
# remove 
# (can delete on value, here no need of index position)
# don't need to find index position
l = [1,2,3,4,5]
l.remove(5)
print(l)

```

```python
# pop
# here have to tell index number, if don't give then delete last 
# element in the list.

l = [1,2,3,4,5]
l.pop(2)
# l.pop()
print(l)
```

```python
# clear 
l = [1,2,3,4,5]
l.clear()
print(l)
```

```python
l1 = [1,2,3,4,5]
l2 = [5,6,7,8,9,[10,11]]

print(5 in l1)
print(10 in l2)
# print(10 in l2[-1]) True
# print([10,11] in l2) True
```

```python
# loops
result = []
for i in l1:
    if i not in l2 and i not in result:
        result.append(i)
        
for i in l2:
    if i not in l1 and i not in result:
        result.append(i)
        
print(result)
```

## List Functions

```python
# len/min/max/sorted
l = [11,1,2,3,4,5]
print(len(l))
print(min(l)) # TypeError: '<' not supported between instances of 'str' and 'int'
print(max(l))
print(sorted(l,reverse=True)) # sorted is not the a parmanent operator
```

```python
# copy -> shallow (naya addess me ek ka address store karta hai)
l = [2,1,5,7,0]
print(l)
print(id(l))
l1 = l.copy()
print(l1)
print(id(l1))
```

## List Comprehension

List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]

Advantages of list Comprehension

* More time-efficient and space-efficient than loops.
* Require fewer lines of code.
* Transforms iterative statement into a formula.


```python
# Add 1 to 10 numbers to a list
l = []

for i in range(1,11):
    l.append(i)

print(l)

```

```python
l = [i for i in range(1,11)]
print(l)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = -3
# [-6,-9,-12]


x = []
for i in v:
    x.append(i*s)
    
print(x)
```

```python
l = [s*i for i in v]
l
```

```python
# Add squares
l = [1,2,3,4]
s = []

for i in l:
    s.append(i**2)
    
print(s)
```

```python
l = [1,2,3,4]
l = [i**2 for i in l]
l
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50
l = [i for i in range(1,51) if i%5==0]
l
```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']

l = [language for language in languages if language.startswith('p')]
l
```

```python
# Nested if with List Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'
# items= []
# for i in my_fruits:
#     if i in basket and i.startswith('a'):
#         items.append(i)
        
# print(items)

# l = [i for i in my_fruits if i in basket and i.startswith('a')]

# using nested list
l = [i for i in my_fruits if i in basket if i.startswith('a')]

l      
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products -> list comprehension on 2 lists together
l1 = [1,2,3,4]
l2 = [5,6,7,8]

[i*j for i in l1 for j in l2]
```
