### Techniques to measure time efficiency



## 1. Meausring time to execute (it's depend upon  machine)

```python
import time
start = time.time() #current time 
for i in range(1,11):
    print(i)
    
end = time.time() 

print(end-start)
```

```python
import time
start = time.time() #current time 
i = 0
while i !=10:
    i+=1
    print(i)
    
end = time.time() 

print(end-start)
```

## Problems with this approach

* Different time for different algorithm -->(it's good)
* Time varies if implementation changes   (don't have to change time after change the implement the code with different ways.)
* Different machines different time       (it's majorly depends on hardware)
* Does not work for extremely small input 
* Time varies for different inputs, but can't establish a relationship. (unable to find mathematical relationship between input and time for execution)

## 3. Orders of Growth

### Goals:
* want to evaluate program's efficiency when input is very big 
* want to express the growth of program's run time as input size grows
* want to put an upper bound on growth - as tight as possible
* don't need to be  precise: "order of" not "exact" growth
* we will look at largest factors in run time (which section of the program will take the longest to run?)
* thus, generally we want tight upper bound on growth, as function of size of input, in worst case.

## What is meant by Big O notation?
Big-O Notation: A simple explanation with examples
Big-O notation is the language we use for talking about how long an algorithm takes to run (time complexity) or how much memory is used by an algorithm (space complexity). Big-O notation can express the best, worst, and average-case running time of an algorithm.
