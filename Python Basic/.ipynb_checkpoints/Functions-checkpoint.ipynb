## *args     *kwargs

*args and **kwargs are special Python keywords that are used to pass the variable length of arguments to a function

*args internally create a tuple and adds the items like (1,2,3,4,5,6,6).

```python
# *args (aisa function banana hai ki kitne bhi number of inputs ko le sake or ouput de.)
# allow to take unlimited input for the function and perform task.
def multiply(a,b):
    return a*b
multiply(2,3)
```

```python
# create a function which don't have essue to how much user throwing inputs. 
def multipy(*args):  # instead of args we can take anything.
    p = 1
    
    for i in args:
        p = p*i
    print(args)
    return p
        
```

```python
multipy(1,2,3,4,5,6,2,3,4,5,6)
```

```python
def multipy(*vishal):  # instead of args we can take anything.
    p = 1
    
    for i in vishal:
        p = p*i
    print(vishal)
    return p
```

```python
multipy(1,2,3,4,5,6,2,3,4,5,6)
```

## **kwargs ==> kw-> keywords

It allows us to pass any number of keyword arguments.
Keyword rguments mean that they contain a key-value pair, like a Python Dictionary.

```python
# **kwargs  ==> beacome a dictionary. like args be a tuple
def display(**d): # we can write anything here after this ** instead of kwargs 
    for (key,value) in d.items():
        print(key,'-->',value)
        
display(india='delhi',srilanka='colombo',nepal='kathmandu',pakistan='islamabad')
```

#### Points to remember while using *args and **kwargs
* order of the arguments matter(normal -> *args -> **kwargs)
* The words "args" and "kwargs" are only a convention, you can use any name of your choice.

```python
# agar function ke under variable nahi hai to global wale ko use kar sakta hai but change nahi kar sakta
# def h(y): UnboundLocalError: cannot access local variable 'x' where it is not associated with a value
#     x += 1
# x = 5
# h(x)
# print(x)
```

```python
# agar apko karna hi hai. to use global
def h(y):
    global x
    x += 1
    print(x)
x = 5
h(x)
print(x)
```

```python
def f(x):
    x = x+1
    print('in f(x): x =',x)
    return x

x = 3
z = f(x)

print('in main program scope: z =',z)
print('in main program scope: x =',x)
```

## Lambda Function
A lambda function is a small anonymous function.(ek chhota function jiska naam nahi hota.)

A lambda function can take any number of arguments, but can only have one expression.

```python
from IPython.display import display
from PIL import Image
img = Image.open("/home/vishal/Pictures/Screenshots/Screenshot from 2024-01-30 16-07-53.png")
display(img)
```

```python
# x -> x^2
a=lambda x:x**2
a(3)
```

```python
# x,y -> x+y
s=lambda x,y:x+y
s(2,3)
```

## Diff between lambda vs Normal Function
* No name (lambda function has not name)
* lambda has no return value (infact, returns a function) value nahi pura ka pura fuction hi return kar deta hai.
* lambda is always written in 1 line.
* not reusable

Then why use lambda functions
They are used with HOF(Higher Order Functions ke sath)



```python
# check if a string has 'a'
a = lambda s:'a' in s
a('vishal')
```

```python
# odd or even
b = lambda n:'even' if n%2==0 else 'odd'
b(2)
```

## Higher Order Functions
a function take as input 

a function jo function ko return kare

it both condition means higher order functions.

```python
# Example

def square(x):
    return x**2

# HOF
def transform(f,l):
    output=[]
    for i in l:
        output.append(f(i))
    print(output)
    
l = [1,2,3,4,5]
transform(square,l)
```

```python
# insted of it use lambda
transform(lambda x:x**3,l)
```

## Map 
(accept lambda function, iterable)

```python
# square the items of a list (we can do type convert on it)
list(map(lambda x:x**2,[1,2,3,4,5]))
```

```python
# odd/even labelling of list items
l = [1,2,3,4,5]
list(map(lambda x:'even' if x%2==0 else 'odd',l))
```

```python
user=[
    {
        'name':'vishal',
        'age':22,
        'gender':'male'
    },
    {
        'name':'rahul',
        'age':25,
        'gender':'male'
    },
    {
        'name':'gaurav',
        'age':27,
        'gender':'male'
    }
]
```

## Filter
ek given list ko kisi given condition se filter karta hai

```python
# numbers greater than 5
l = [3,4,5,6,6,6,7,8,8,6,7]

list(filter(lambda x:x>5,l))

# map har number pe logic applay karta hai or true false dikata hai.
# filter true value show karta hai
```

```python
# fetch fruits starting with 'a'

fruits = ['apple','guava','cherry','mango','ananas']

list(filter(lambda x:x.startswith('a'),fruits))
```

## Reduce

```python
# sum of all item
import functools

functools.reduce(lambda x,y:x+y,[1,2,3,4,5])
```

```python
# find min
functools.reduce(lambda x,y:x if x<y else y,[23,11,45,10,1])
```
